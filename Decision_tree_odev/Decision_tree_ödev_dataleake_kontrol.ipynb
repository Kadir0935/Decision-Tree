```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("nyc-east-river-bicycle-counts.csv")
```

```python
# 1. Her şeyi stringe çevir
df['Precipitation'] = df['Precipitation'].astype(str)

# 2. (S) gibi parantez içindeki her şeyi kaldır (örneğin "0.47 (S)" → "0.47")
df['Precipitation'] = df['Precipitation'].str.replace(r'\(.*\)', '', regex=True)

# 3. 'T' değerini çok küçük yağış olarak 0.01 yap
df['Precipitation'] = df['Precipitation'].replace('T', '0.01')

# 4. sayıya çevir
df['Precipitation'] = df['Precipitation'].astype(float)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['DayOfWeek'] = df['Date'].dt.dayofweek
df = df.drop('Day', axis=1)
```

```python
df.describe()
```

```python
# Yağış vs Toplam trafik
sns.scatterplot(data=df, x='Precipitation', y='Total')
plt.title('Yağışın Toplam Trafiğe Etkisi')
plt.show()

# Sıcaklık vs Toplam trafik
sns.scatterplot(data=df, x='High Temp (°F)', y='Total')
plt.show()
```

```python
# Köprü bazında trafik dağılımı
sns.boxplot(data=df[['Brooklyn Bridge','Manhattan Bridge','Williamsburg Bridge','Queensboro Bridge']])
plt.title('Köprü Trafik Dağılımı')
plt.show()
```

```python
drop_cols = ['Brooklyn Bridge','Manhattan Bridge','Williamsburg Bridge','Queensboro Bridge','Unnamed: 0','Date']
df_model = df.drop(columns=drop_cols)
```

```python
df_model.columns
```

```python
X = df_model.drop(["Total"],axis=1)
y = df_model["Total"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=15)
```

```python
X_train.isnull().sum()
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
linear = LinearRegression()
tree = DecisionTreeRegressor(
    max_depth=5,          
    min_samples_split=10, 
    min_samples_leaf=5,    
    random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
linear.fit(X_train_scaled, y_train)
```

```python
y_pred = linear.predict(X_test_scaled)
```

```python
check_data=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
check_data[:15]
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print("r2 score: ", r2_score(y_test,y_pred))
print("mean absolute error: ", mean_absolute_error(y_test,y_pred))
print("mean squared error: ", mean_squared_error(y_test,y_pred))
```

```python
tree.fit(X_train_scaled, y_train)
```

```python
y_pred=tree.predict(X_test_scaled)
```

```python
print("r2 score: ", r2_score(y_test,y_pred))
print("mean absolute error: ", mean_absolute_error(y_test,y_pred))
print("mean squared error: ", mean_squared_error(y_test,y_pred))
```

```python
pd.DataFrame({
    'Feature': X.columns,
    'Importance': tree.feature_importances_
}).sort_values(by='Importance', ascending=False)
```

```python
from sklearn.model_selection import cross_val_score, KFold
```

```python
# 5-fold cross validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Decision Tree için CV
dt_scores = cross_val_score(DecisionTreeRegressor(), X, y, 
                           cv=kfold, scoring='r2')
print(f"Decision Tree CV R²: {dt_scores.mean():.3f} (±{dt_scores.std():.3f})")

# Linear Regression için CV
lr_scores = cross_val_score(LinearRegression(), X, y, 
                           cv=kfold, scoring='r2')
print(f"Linear Regression CV R²: {lr_scores.mean():.3f} (±{lr_scores.std():.3f})")
```

```python
# 1. Veri boyutunu kontrol et
print(f"Veri şekli: {X.shape}")

# 2. Data leakage kontrolü
# - Feature'lar target'tan bağımsız mı?
# - Zaman bilgisi var mı?
# - Yanlış feature engineering?

# 3. Decision Tree derinliğini kontrol et
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)
print(f"Tree depth: {dt.get_depth()}")
print(f"Number of leaves: {dt.get_n_leaves()}")

# 4. Train vs Test performansı
train_score = dt.score(X_train, y_train)
test_score = dt.score(X_test, y_test)
print(f"Train R²: {train_score:.3f}")
print(f"Test R²: {test_score:.3f}")
```

```python
simple_dt = DecisionTreeRegressor(max_depth=2, random_state=42)
simple_dt.fit(X_train, y_train)

simple_train_score = simple_dt.score(X_train, y_train)
simple_test_score = simple_dt.score(X_test, y_test)

print(f"Basit Tree Train R²: {simple_train_score:.3f}")
print(f"Basit Tree Test R²: {simple_test_score:.3f}")
```

```python
dt_optimized = DecisionTreeRegressor(
    max_depth=4, 
    min_samples_split=20, 
    min_samples_leaf=10,  
    max_features=0.8,      
    random_state=42
)

dt_optimized.fit(X_train, y_train)
train_score_opt = dt_optimized.score(X_train, y_train)
test_score_opt = dt_optimized.score(X_test, y_test)
print(f"Optimized Tree - Train: {train_score_opt:.3f}, Test: {test_score_opt:.3f}")
```

```python
final_model = DecisionTreeRegressor(
    max_depth=4,
    min_samples_split=20,
    min_samples_leaf=10,
    random_state=42
)

final_model.fit(X_train, y_train)


predictions = final_model.predict(X_test)
```

```python
print("r2 score: ", r2_score(y_test,predictions))
print("mean absolute error: ", mean_absolute_error(y_test,predictions))
print("mean squared error: ", mean_squared_error(y_test,predictions))
```
